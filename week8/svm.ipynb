```python
# 1. Import necessary libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score, classification_report,confusion_matrix
```

```python
# 19. Prepare x and y

x = df.drop(columns="Species")
y = df["Species"]

print(x.head())
print(y.head())
```

```python
# 20. split train and test to 70% train data, random state = 1

x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.7,random_state=1)
```

```python
# 21 - 24. Model building : Train SVC model using cost C = 1

model = SVC(C=1)
model.fit(x_train,y_train)
```

```python
# 25. Predict using the model

y_pred = model.predict(x_test)

y_pred[:5]
```

```python
# 26. Evaluate the model using confusion matrix

print("Confusion matrix : \n" , confusion_matrix(y_test,y_pred))
```

```python
#  27 - 28. Print classification report and accuracy
print("Classification report : \n", classification_report(y_test,y_pred))
print(f"Accuaracy score : {accuracy_score(y_test,y_pred):.2f}")
```

```python
# 29. K-fold cross validation

scores = cross_val_score(model, x, y, cv=5)  # 5-fold CV
print("Cross-validation scores:", scores)
print("Average accuracy:", scores.mean())
```

```python
# 30. Use GridSearchCV to find optimal value of C
# 31. Try with C = [0.1, 1, 10, 100, 1000]
# 32. Specify number of folds for K-Fold CV


param_grid = {'C': [0.1, 1, 10, 100, 1000]}
grid = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy', return_train_score=True)
grid.fit(x_train, y_train)
print("Best parameters:", grid.best_params_)
```

```python
# 33. Fit best model on the training data

best_model = grid.best_estimator_
y_pred = best_model.predict(x_test)

print("Classification report : \n", classification_report(y_test,y_pred))
print(f"Accuaracy score : {accuracy_score(y_test,y_pred):.2f}")
```

```python
# 34. Print the scores of grid search cv as data frame

results_df = pd.DataFrame(grid.cv_results_)
print(results_df[['param_C', "mean_train_score", 'mean_test_score', 'rank_test_score',]])
```

```python
# 35. Plot of c vs train and test scores

test_scores = results_df['mean_test_score']
train_scores = results_df['mean_train_score']


plt.figure(figsize=(10, 6))
plt.plot(param_grid['C'], test_scores, label='Test Score', marker='o')
plt.plot(param_grid['C'], train_scores, label='Train Score', marker='x')
plt.xlabel('C Value')
plt.ylabel('Accuracy')
plt.title('C vs Accuracy')
plt.legend()
plt.show()
```

```python
# 36. Print best score and best C value

print("Best C value:", grid.best_params_['C'])
print("Best cross-validated accuracy:", grid.best_score_)
```
